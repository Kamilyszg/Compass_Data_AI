9. Nota-se que não há tanta variação dos dados em relação à média da coluna, então a substituição dos valores nulos pelo valor da média é uma boa prática que garante a consisteância das informações.

```python
#substituindo os valores NaN pela média
media_rating = df_sem_duplicatas['Rating'].mean()
df_sem_duplicatas['Rating'] = df_sem_duplicatas['Rating'].fillna(media_rating)
```

13. Considerando que as duplicatas diferem-se pelo número de reviews que aumenta conforme a passagem do tempo, faz-se necessária a conversão da coluna em um tipo numérico, assim pode-se consultar qual é o valor máximo por app e selecionar somente este (o mais recente) para que prevaleça no dataframe

```python
#convertendo a coluna reviews para numérica
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')
df.info()
```

14. Remoção das versões antigas dos aplicativos

```python
#ordenação por reviews de maneira decrescente
df_sorted = df.sort_values(by='Reviews', ascending=False)
#remoção das duplicatas considerando a coluna App, deixando somente a primeira ocorrência (com mais reviews)
df_sem_duplicatas = df_sorted.drop_duplicates(subset='App', keep='first')
df_sem_duplicatas
```

## Gráfico 1: Top 5 Apps por número de instalação

1. Em primeiro lugar, é necessário converter o tipo da coluna 'Installs' para inteiro a fim de que os dados possam ser tratados como tal

```python
#Criação de função para a conversão
def convert_installs(value):
    #retirando as vírgulas dos números e o símbolo +
    value = value.replace(',' , '').replace('+' , '')
    return int(value)

df_sem_duplicatas['Installs'] = df_sem_duplicatas['Installs'].apply(convert_installs)
df_sem_duplicatas['Installs']
```

2. Ordenação pelos valores da coluna 'Installs'

```python
df_sorted = df_sem_duplicatas.sort_values(by='Installs', ascending= False)
df_sorted.head()
```

3. Importação da biblioteca mathplotlib para a construção do gráfico de barras vertical

```python
import matplotlib.pyplot as plt
```

4. Atribuição da lista dos mais instalados a uma variável a parte

```python
mais_instalados = df_sorted.head()
```

5. Criação do gráfico de acordo com a documentação da biblioteca

```python
apps = mais_instalados['App']
installs = mais_instalados['Installs']

#criação da figura e do eixo que será construído o gráfico
fig, ax = plt.subplots()

#definindo o tipo de gráfico (barras) e as informações que estarão em x e y, além das cores das barras
ax.bar(apps, installs, color=['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple'])

#definindo labels do eixo y
ax.set_ylabel('Instalações (mi)')
#título do gráfico
ax.set_title('Top 5 Aplicativos Mais Instalados')

#rotacionando os títulos para melhorar a legibilidade 
plt.xticks(rotation=45, ha='right')  

#ajuste do layout
plt.tight_layout()

#salvando o gráfico como png
plt.savefig('top5.png')

#exibindo
plt.show()
```

## Gráfico 2: Frequência das categorias existentes

1. Cálculo da frequência de cada categoria

```python
#contabiliza a quantidade de vezes que aparece a categoria com value_counts
#normalize true normaliza os dados para que a soma das frequências seja 1
# * 100 - conversão em porcentagem
categoria_frequencia = df['Category'].value_counts(normalize=True) * 100
print(categoria_frequencia)
```

2. Para melhor visualização das informações no gráfico a ser construído, as categorias que possuem uma frequência menor de 1% serão agrupadas em 'Outros'

```python
# Separando as categorias que tem menos de 1% e agrupá-las como outros
menor_que_1 = categoria_frequencia[categoria_frequencia < 1]
categoria_frequencia = categoria_frequencia[categoria_frequencia >= 1]
```

```python
# Somando as porcentagens de categorias menores que 1% 
categoria_frequencia['OTHERS'] = menor_que_1.sum()
```

3. Criação do gráfico 'Pie' de acordo com a documentação apresentada

```python
#criação da figura, do eixo e definindo seu tamanho
fig, ax = plt.subplots(figsize=(12, 8))

ax.pie(categoria_frequencia,
        labels=categoria_frequencia.index, #passagem do index das categorias para labels
        autopct='%1.1f%%', #formato das porcentagens
        startangle=90, #ângulo inicial
        pctdistance=0.9 #distância entre porcentagens
        )

# Ajustando layout circular
ax.axis('equal')

# Adicionando título
ax.set_title('Frequência das Categorias de Aplicativos')
# Ajustando o tamanho da fonte das labels
plt.setp(ax.texts, fontsize=8) 

#salvando como png
plt.savefig('freq_categorias.png')

# Exibindo o gráfico
plt.show()
```
